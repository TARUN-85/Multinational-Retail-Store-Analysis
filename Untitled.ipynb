#### 3.2 Bivariate Analysis

```python
# Before removing outliers values
print("Before removing outliers values")
plt.figure(figsize=(20,5))
plt.title('Product Category Purchased by customer Gender wise at different costs')
sns.boxplot(data=df,x='Product_Category',y='Purchase',hue='Gender')
plt.show()


# After removing outliers values
print("After removing outliers values")
a = df[['Product_Category','Purchase','Gender']].copy()
for i in range(1,21):
    #Removing Male outliers
    b = a.loc[(a['Product_Category']==i) & (a['Gender']=='M'),'Purchase']
    p_25,p_75 = b.quantile([0.25,0.75])
    iqr = p_75 - p_25
    mx = p_75 + iqr*1.5
    mn = p_25 - iqr*1.5
    a.drop(index=a[ ((a['Gender']=='M') & a['Product_Category']==i) & ((a['Purchase']> mx) | (a['Purchase'] < mn))]
           .index,inplace=True)
    
    #Removing Female outliers
    c = a.loc[(a['Product_Category']==i) & (a['Gender']=='F'),'Purchase']
    p_25,p_75 = c.quantile([0.25,0.75])
    iqr = p_75 - p_25
    mx = p_75 + iqr*1.5
    mn = p_25 - iqr*1.5
    a.drop(index=a[ ((a['Gender']=='F') & a['Product_Category']==i) & ((a['Purchase']> mx) | (a['Purchase'] < mn))].index,inplace=True)


plt.figure(figsize=(20,5))
sns.boxplot(data=a,x='Product_Category',y='Purchase',hue='Gender')
plt.show()
```

##  5.1 for Gender & Purchase

### Calculation of population mean(mu) & Standard-Deviation (sd) for Gender & Purchase

```python
# Male Population Calculations
a = df.loc[df['Gender']=='M','Purchase']
male_mu = a.mean()
male_sd = a.std()
print("The population Mean is mu:{} & population Standard Deviation is sd:{} when gender is Male".format(male_mu,male_sd))

# Female Population Calculations
a = df.loc[df['Gender']=='F','Purchase']
female_mu = a.mean()
female_sd = a.std()
print("The population Mean is mu:{} & population Standard Deviation is sd:{} when gender is Female".format(female_mu,female_sd))
```

###  Calculation for (300,3000,3000) sample size & Different confident (alpha = 1%,5%,10% respectively) combinations

```python
print("For Sample size of 300 & 99% Confidence i.e. (alpha = 1%)")
sample_size = 300
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=300) & 99% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_300_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 300 samples :",Male_sample_300_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_300_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 300 samples :",Female_sample_300_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 1/100
Z_300 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 99% confidence is [{}, {}]"
      .format(male_mu - Z_300 *Male_sample_300_sd,
              male_mu + Z_300 *Male_sample_300_sd))

# Range Calculation for Female people
print("CI range for Female at 99% confidence is [{}, {}]"
      .format(female_mu - Z_300 *Female_sample_300_sd,
              female_mu + Z_300 *Female_sample_300_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 300 & 95% Confidence i.e. (alpha = 5%)")
sample_size = 300
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=300) & 95% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_300_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 300 samples :",Male_sample_300_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_300_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 300 samples :",Female_sample_300_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 5/100
Z_300 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 95% confidence is [{}, {}]"
      .format(male_mu - Z_300 *Male_sample_300_sd,
              male_mu + Z_300 *Male_sample_300_sd))

# Range Calculation for Female people
print("CI range for Female at 95% confidence is [{}, {}]"
      .format(female_mu - Z_300 *Female_sample_300_sd,
              female_mu + Z_300 *Female_sample_300_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 300 & 90% Confidence i.e. (alpha = 10%)")
sample_size = 300
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=300) & 90% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_300_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 300 samples :",Male_sample_300_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_300_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 300 samples :",Female_sample_300_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 10/100
Z_300 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 90% confidence is [{}, {}]"
      .format(male_mu - Z_300 *Male_sample_300_sd,
              male_mu + Z_300 *Male_sample_300_sd))

# Range Calculation for Female people
print("CI range for Female at 90% confidence is [{}, {}]"
      .format(female_mu - Z_300 *Female_sample_300_sd,
              female_mu + Z_300 *Female_sample_300_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 3000 & 99% Confidence i.e. (alpha = 1%)")
sample_size = 3000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=3000) & 99% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_3000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 3000 samples :",Male_sample_3000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_3000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 3000 samples :",Female_sample_3000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 1/100
Z_3000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 99% confidence is [{}, {}]"
      .format(male_mu - Z_3000 *Male_sample_3000_sd,
              male_mu + Z_3000 *Male_sample_3000_sd))

# Range Calculation for FemaleFemale people
print("CI range for Female at 99% confidence is [{}, {}]"
      .format(female_mu - Z_3000 *Female_sample_3000_sd,
              female_mu + Z_3000 *Female_sample_3000_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 3000 & 95% Confidence i.e. (alpha = 5%)")
sample_size = 3000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=3000) & 95% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_3000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 3000 samples :",Male_sample_3000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_3000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 3000 samples :",Female_sample_3000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 5/100
Z_3000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 95% confidence is [{}, {}]"
      .format(male_mu - Z_3000 *Male_sample_3000_sd,
              male_mu + Z_3000 *Male_sample_3000_sd))

# Range Calculation for Female people
print("CI range for Female at 95% confidence is [{}, {}]"
      .format(female_mu - Z_3000 *Female_sample_3000_sd,
              female_mu + Z_3000 *Female_sample_3000_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 3000 & 90% Confidence i.e. (alpha = 10%)")
sample_size = 3000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=3000) & 90% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_3000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 3000 samples :",Male_sample_3000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_3000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 3000 samples :",Female_sample_3000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 10/100
Z_3000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 90% confidence is [{}, {}]"
      .format(male_mu - Z_3000 *Male_sample_3000_sd,
              male_mu + Z_3000 *Male_sample_3000_sd))

# Range Calculation for Female people
print("CI range for Female at 90% confidence is [{}, {}]"
      .format(female_mu - Z_3000 *Female_sample_3000_sd,
              female_mu + Z_3000 *Female_sample_3000_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 30000 & 99% Confidence i.e. (alpha = 1%)")
sample_size = 30000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=30000) & 99% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_30000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 30000 samples :",Male_sample_30000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_30000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 30000 samples :",Female_sample_30000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 1/100
Z_30000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 99% confidence is [{}, {}]"
      .format(male_mu - Z_30000 *Male_sample_30000_sd,
              male_mu + Z_30000 *Male_sample_30000_sd))

# Range Calculation for Female people
print("CI range for Female at 99% confidence is [{}, {}]"
      .format(female_mu - Z_30000 *Female_sample_30000_sd,
              female_mu + Z_30000 *Female_sample_30000_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 30000 & 95% Confidence i.e. (alpha = 5%)")
sample_size = 30000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=30000) & 95% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_30000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 30000 samples :",Male_sample_30000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_30000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 30000 samples :",Female_sample_30000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 5/100
Z_30000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 95% confidence is [{}, {}]"
      .format(male_mu - Z_30000 *Male_sample_30000_sd,
              male_mu + Z_30000 *Male_sample_30000_sd))

# Range Calculation for Female people
print("CI range for Female at 95% confidence is [{}, {}]"
      .format(female_mu - Z_30000 *Female_sample_30000_sd,
              female_mu + Z_30000 *Female_sample_30000_sd))





print()
print("_ "*60)

##############################################################################################################################

print("For Sample size of 30000 & 90% Confidence i.e. (alpha = 10%)")
sample_size = 30000
Male_Sample_Purchase_Mean = []
Female_Sample_Purchase_Mean = []
for i in range(1,1000):
    Male_Mean = df.loc[df['Gender']=='M','Purchase'].sample(sample_size).mean()
    Male_Sample_Purchase_Mean.append(Male_Mean)
    Female_Mean = df.loc[df['Gender']=='F','Purchase'].sample(sample_size).mean()
    Female_Sample_Purchase_Mean.append(Female_Mean)
# Plotting histogram for both Male & Female for Purchases
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.histplot(Male_Sample_Purchase_Mean,kde=True)
plt.title("Male vs Purchase")

plt.subplot(1,2,2)
sns.histplot(Female_Sample_Purchase_Mean,kde=True)
plt.title("Female vs Purchase")
plt.show()

############################ for sample(n=30000) & 90% confidence calculation of CI ##########################################
# Male Sample mean Calculation & Standar deviation Calculation for Sample
Male_sample_30000_sd = male_sd/math.sqrt(sample_size)
print("Standard Deviation for Male of 30000 samples :",Male_sample_30000_sd)

#Female Sample mean Calculation & Standar deviation Calculation for Sample
Female_sample_30000_sd = female_sd/math.sqrt(sample_size)
print("Standard Deviation for Female of 30000 samples :",Female_sample_30000_sd)

############# Calculation of Z-Score & confidence Interval (CI = [mu - Z*(s.d/root(n)), mu - Z*(s.d/root(n))]) ################
alpha = 10/100
Z_30000 = abs(norm.ppf(alpha))

# Range Calculation for Male Gender
print("CI range for Male at 90% confidence is [{}, {}]"
      .format(male_mu - Z_30000 *Male_sample_30000_sd,
              male_mu + Z_30000 *Male_sample_30000_sd))

# Range Calculation for Female people
print("CI range for Female at 90% confidence is [{}, {}]"
      .format(female_mu - Z_30000 *Female_sample_30000_sd,
              female_mu + Z_30000 *Female_sample_30000_sd))





print()
print("_ "*60)

##############################################################################################################################
```
